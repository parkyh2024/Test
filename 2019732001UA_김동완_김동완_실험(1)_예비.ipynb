# $$머신러닝이란$$
***

##1. 개요
머신러닝(Machine Learning), 즉 **기계학습**은 컴퓨터를 인간처럼 학습시켜 스스로 규칙을 형성하도록 한 기술로 경험적 데이터를 기반으로 학습을 하고 예측을 수행하고 그를 통해서 스스로 성능을 향상시키는 시스템과 이를 위해 알고리즘을 연구하고 구축하는 기술이다. 머신러닝을 하는 방법에는 크게 3가지가 있다.
***

##2. 학습방법에 따른 분류
![인공지능 분류 사진](https://hyeonjiwon.github.io/assets/img/21-3.png)

####1. **지도학습(Supervised Learning)**  
문제와 정답을 모두 알려준 뒤 학습을 하도록 하는 방법으로 그 다음 데이터에 대해 예측하고자 하는 값을 올바르게 추측하고 분류하도록 하는 학습 방법이다.  
입력과 출력쌍이 주어지는 경우 그것들 사이의 대응관계를 학습한다.
####2. 비지도 학습(Unsupervised Learning)  
답을 가르쳐주지 않고 스스로 학습을 하도록 하는 방법으로 입력값만 있는 훈련 데이터를 이용한다.  
입력 사이의 연관성이나 규칙성을 스스로 찾아내도록 하는 학습을 한다.
####3. 강화 학습(Reinforcement Learning)  
행동의 좋고 나쁜 정도를 알려주어 주어진 상황에서 최적의 행동을 선택 또는 강화 하도록 하는 학습 방법이다.  
최대의 보상을 목표로 학습하도록 하여 최적의 행동이나 행동순서를 선택하게 한다.  

||머신러닝||
|:---:|:---:|:---:|
|지도학습|비지도학습|강화학습|

##3. 지도학습(Supervised Learning)
주로 사용 할 머신러닝의 학습 방법으로 다음과 같은 과정으로 학습이 이루어 진다  
![지도학습(Supervised Learning) 예시](https://static.wixstatic.com/media/c176d5_e0565d75109e424eac79f097e1c9ee2b~mv2.png/v1/fill/w_773,h_482,al_c,q_90,enc_auto/c176d5_e0565d75109e424eac79f097e1c9ee2b~mv2.png)  
지도학습은 이미 학습된 데이터를 기반으로 새로운 데이터를 분류하고 그 데이터가 무엇인지 정확하게 예측하는 과정으로 이루어진다.  
지도 학습은 크게 **분류기법**과 **선형 회귀분석** 의 두 가지의 유형으로 분류할 수 있다.  
분류기법은 전에 학습했던 것을 토대로 새로운 데이터를 알맞게 분류하는 것으로 이미지 인식과 같은 용도에 적합하다.  
선형 회귀분석은 데이터간의 관계를 학습하여 예측하도록 하는 방식이다.
