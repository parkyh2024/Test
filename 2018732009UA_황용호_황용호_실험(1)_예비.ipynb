###1. **[머신러닝](https://terms.naver.com/entry.naver?docId=3347329&cid=40942&categoryId=32845)**이란? 
###### : 인간의 학습 능력과 같은 기능을 컴퓨터에서 실현하고자 하는 기술을 의미, 세가지로 분류 가능
* 지도 학습 (Supervised Learning) : 정답을 <u>알고 있는</u> 문제를 풀게 하는 방법  
* 비지도 학습 (Unsupervised Learning) : 정답을 <u>모르는</u> 문제를 풀게 하는 방법  
* 강화 학습 (Reinforcement Learning) : 문제를 맞출 경우 <u>큰 보상</u>, 틀린 경우 <u>손해를</u> 통해 피드백 하는 방법  
이 중에서 **지도 학습**에 중점을 두고 실험을 할 예정  

###2. **지도 학습** (Supervised Learning)이란 ? 
 >  -> 전문가로 부터 준비된 <u>정답을 받고</u>  
 >  -> 반복 학습을 통한 <u>이해</u>  
 >  -> 새로운 데이터에 대한 <u>정답을 제시</u>   
 >  한계점 : 배우지 않은 것은 알 수 없음  

###3. **Google Colab**란? 
######: 구글의 머신러닝 클라우드 플랫폼
>1) 이용방법 :   
>>(1) <https://colab.research.google.com/> 들어가기  
>>(2) 새 노트 (Untitled0.ipynb)생성 후 구글 드라이브와 연결   
>>(3) 폴더 - 드라이브 마운트 선택   
>>(4) Google Drive에 연결 선택   
>>(5) drive 폴더 생성   
>>(6) drive - MyDrive - Colab Notebooks - Untitled0.ipynb  
>>(7) 런타임 - 런타임 유형 변경 선택  
>>(8) 하드웨어 가속기 GPU로 설정 후 저장   
>>(9) 설정 완료 후 코드 입력이나 텍스트 입력 가능 

>2) 코드 입력  
: 파이썬 코드 입력 후 실행 가능, 텍스트를 통해 코드 설명 가능.

>3) Notebook 단축기  
: 주로 '실행 후 다음 셀로 이동(Shift+Enter)', 'Markdown 전환(ctrl+MM)', '코드 전환(Ctrl+MY)', '코드 셀에 줄번호 부여(Ctrl+ML)'을 많이 사용함. 


###4. **Markdown**란?
######: 서식 편집 언어를 코딩을 이용해 문서 작성함
>1) 타이틀 : #개수에 따라 문자 크기 조정   
>>EX) #(가장 큰)~#######(작은)  

>2) 강조 : '진하게','밑줄', '취소', '이탤릭'   
>>EX) '진하게' : '`**(문자)**`' -> **황용호**  
  EX) '밑줄' : '`<u>(문자)</u>`' -> <u>황용호</u>  
  EX) '취소' : '`~~(문자)~~`' -> ~~황용호~~  
  EX) '이탤릭' : '`_(문자)_`' -> _황용호_

>3) 줄바꿈 : 문장의 끝에서 'Space' 2번 후 'Enter'  
>4) 블럭(인용) 들여쓰기 : `>(문자)`= 1단계, `>>(문자)`=2단계  
>5) 목록 : `'*', '+', '-'` 세가지 모두 사용 가능   
>>'tap'으로 들여쓰기 후 사용 시 문자가 바뀜.  
>> 숫자 사용시에는 `(숫자).`으로 쓰면 자동으로 인식  

>6) Code Block : ' ` '사용해서 'C언어'나 '파이썬'으로 디자인 가능  
>7) 하이퍼링크 : 한글로 표시 방법과 바로 링크 주소로도 가능  
>>EX) `[광운대 로그인 페이지 바로가기](https://klas.kw.ac.kr/)` -> [광운대 로그인 페이지 바로가기](https://klas.kw.ac.kr/)   
  EX) `<https://klas.kw.ac.kr/>`-> <https://klas.kw.ac.kr/>

>8) 수평선 : ' * '과 ' - '를 세개 이상 사용시 수평선이 나타남   
>9) 표 : 내용이 큰 경우 직접 작성하기 보단 웹페이지 활용  
>10) 그림 : `![이미지의 설명](이미지의 주소)` 사용 시 출력결과 이미지를 불러옴  
>11) 체크박스 : 목록화 후  `'[ ]'`사용 시 체크 박스 형태로 나타남, 체크 필요시 `'[x]'`를 이용  
>12) 수식 : `'$$ ~ $$'` 사이에 Latex 문법을 이용해서 표현함
