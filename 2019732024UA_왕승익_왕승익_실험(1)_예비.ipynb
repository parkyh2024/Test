## 3. 블럭(인용) 들여쓰기, 목록, Code Block
>### 3-1. 블럭(인용 들여쓰기)
`>`의 수로 여백 결정  
예시)
> > 한 개 사용
>>> > 세 개 사용

> ### 3-2. 목록
>> 순서 표시가 있는 리스트  
항목 앞에 `1.`, `2.` ...를 입력한다.  
1. 첫번째 리스트 항목입니다.
2. 두번째 리스트 항목입니다.  

>>한 가지 주의할 점은, 순서를 표시하기 위해 숫자를 다르게 입력해도 내림차순으로만 표시됩니다.

>> 순서 표시가 없는 리스트
항목 앞에 `*`, `-`, `+`를 입력합니다. 모두 같은 기호로 입력해도 되고, 기호를 혼합하여 사용해도 됩니다.  

`Tab`을 활용하여 항목의 세부 리스트를 작성 할 수 있습니다.  
예시)  
* 사과
* 오렌지
  * 엄마
  * 아빠

> ### 3-3. Code Block
>>` ``` ~~ ``` `를 코드 전후에 삽입한다.  
예시)  
``` python
import something
a=10
print (a)
if a>10:
  print(a)
```
___

## 4. 하이퍼링크, 수평선, 표
>### 4-1. 하이머링크
`[](https://~.com)`의 형식으로 하이퍼링크를 삽입할 수 있다.  

>### 4-2. 수평선
`___`처럼 언더바를 세개 이상 연속으로 사용하여 수평선을 삽입할 수 있다.  

>### 4-3. 표
`|` 표시로 칸을 나눌 수 있다. 
|제목 셀1|제목 셀2|제목 셀3|제목 셀4|  
|---|:---|---:|:---:|  
|내용 1|내용 2|내용 3|내용 4|  
|내용 5|내용 6|내용 7|내용 8|  

>위의 방법으로 표를 구현할 수 있는데, 첫 줄에 제목을 적어주고, 둘째 줄은 제목과 내용을 구분해주는 역할을 한다.  
`:`을 사용하여 `-`의 좌측에 `:`이 위치하면 왼쪽 정렬, 우측에 위치하면 우측 정렬, 양쪽에 위치하면 가운데 정렬이 가능하다.

google colab에서는 지원하지 않는 것 같다.
___

## 5. 그림, 체크박스
>### 5-1. 그림
`![이미지설명문구](https://그림주소.com)`로 하이퍼링크를 넣는 방식과 유사하지만 앞에 `!`를 붙인다.

>### 5-2. 체크박스
`* [ ]`를 사용하여 비어있는 체크박스를,  
`* [x]`를 사용하여 체크된 체크박스를 구현할 수 있다.
___


## 6. 수식
TeX 문법을 활용하여 수식을 입력할 수 있다.
`$$`를 수식 앞 뒤에 붙여 Code block과 유사한 방법으로 사용한다.
$$y=x^2+\frac{1}{x^4+1}$$

문장 내에 `$ ~ $`를 삽입하여 문장 내에 수식을 삽입할 수도 있다. ($a+7=20$ 같이 말이다.) 많이 쓰이는 표현 중에는 루트를 `\sqrt`로, a부터 b까지 정적분을 하는 경우
 `$\displaystyle \int_a^b f(x) \,{\rm d}x$` 같은 문법을 사용한다.
