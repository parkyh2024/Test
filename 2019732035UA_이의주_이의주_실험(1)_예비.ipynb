### 전기공학머신러닝실습 예비레포트
#### 2019732035 이의주


----
 이번 시간에는 앞으로 머신러닝을 다루기 위해 사용할 코랩에 대해 알아보았다. 먼저 구글 코랩 (Google Colab) 을 사용하기 위해서 https://colab.research.google.com/ 사이트에 접속하여 구글 로그인을 진행하였다. 그리고 새 노트 버튼을 누르며 Untitled0.ipynb 이름의 파일을 생성하였다.  
 이렇게 생성하고 작업한 파일을 저장하는 방법으로는 구글 드라이브에 연결 (Mount)하는 방법이 있다. '노트북이 Google Drive 파일에 엑세스 하도록 허용하시겠습니까?' 라는 문구와 함께 'Google Drive에 연결'을 선택하면 내가 작업한 파일을 드라이브에 저장할 수 있게 된다.  
 그리고 GPU를 사용할 수 있도록 지원해주는 기능이 있는데, 런타임 유형을 변경하면서 하드웨어 가속기에서 GPU를 선택하면서 얻을 수 있다.  
 노트북 파일에서는 코드입력과 텍스트 추가를 할 수 있다. 먼저 코드입력은 단축키 '* /Ctrl+ M B' 또는 +코드 버튼을 눌러 실행할 수 있다. 그리고 그에 따른 설명을 +텍스트 버튼을 누르면서 텍스트 역시 추가할 수 있는 것이다.  
 그리고 이 노트북을 작업하면서 편리하고 더 빠르게 작업을 진행할 수 있도록 Notebook 단축키들이 존재한다. 실행 후 다음 셀 추가는 'Alt + Enter' , 마크다운으로 전환은 'Ctrl+ M M', 실행 멈춤은 'Ctrl + M I' 등을 예시로 들 수 있다.  
 그리고 이 노트북을 작업하러 들어오기 위해 앞서 나온 구글 코랩 사이트로 접속하는 것보다 구글 드라이브에 바로 접속하여 구글 코랩 만들기를 형성하는 것이 더 편리할 수 있다는 점을 알고 있으면 좋을 것 같다.  
 아까 나온 텍스트 파일을 편집하는 기법인 Markdown 사용법의 기초도 알아보았다. Markdown이란 확장자명이 .md인 파일을 말하며, 서식 편집을 언어에서 coding으로 문서를 작성한다는 것이 특징이다. 타이핑만으로 손쉽게 작성할 수 있고, 작성하면서 보이는 문서를 바로 옆에서 확인할 수 있다는 장점을 가지고 있다. Notebook, Blog, Github, Notion 등에서 사용할 수 있어 호환성 역시 좋다고 말할 수 있다.  
 타이틀과 본문을 구분하기 위해서 '#'을 추가하며 글자의 크기를 달리 할 수 있고, 강조하는 부분을 위해서 '-', '~' 등을 사용해 밑줄을 긋거나 취소를 할 수 있다. 예외적으로 밑줄을 그을 때는 'u'를 사용하는 방법이 더 잘 적용된다고 한다. 그리고 블럭(인용) 들여쓰기에 관해서는 '>'을 사용, 목록 추가에는 '*', '+', '-' 등을 사용한다. 이 외에도 하이퍼링크를 달고 수평선, 표, 그림, 체크박스, 수식 등을 추가하며 가독성을 높이는 데 도움을 줄 것이다.  
 이번 강의에서는 이렇게 앞으로 실험을 진행함에 있어 가장 중요한 코랩에 대해서 기초적인 부분을 익힐 수 있었다. 코딩과 그에 대한 설명을 추가할 수 있는 텍스트 작업이 가능한 노트북 파일과 그에 대한 단축키, Markdown 기법까지 확인하였고, 이제 이에 대한 이해도와 숙련도를 높이는 것이 필요하다고 생각한다.
