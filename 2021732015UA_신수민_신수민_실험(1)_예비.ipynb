### **Notebook 단축키 알기**
* 실행된 셀을 선택 : Ctrl + Enter
* 실행된 후 다음 셀로 이동 : Shift + Enter
* 실행 후 다음 셀 추가 : Alt + Enter
* 저장 : Ctrl + S
* 엔터키 : 편집모드
* ESC : 선택모드
* 마크다운으로 전환 : Ctrl + M M
* 코드로 전환 : Ctrl + M Y
* 코드셀에 줄번호 부여 : Ctrl + M L
*바로 윗줄에 셀 생성 : Ctrl + M B
*셀 삭제 : Ctrl + M D

### **타이틀, 강조, 들여쓰기, 목록, Code Block, 하이퍼링크, 수평선**
####**타이틀**  
'#' 을 통해 타이틀을 작성한다. 이때 '#'의 개수가 많을수록 글자가 작아진다. 밑은 이에 해당하는 출력 결과이다.
#TITLE 1
##TITLE 2
###TITLE 3
####Title 4

####**강조**
'**', '~~', ''_' 를 통해 나타내고자 하는 내용을 진하게, 취소 그리고 이탤릭처럼 나타낼 수 있다. 아래는 이를 나타낸 출력 결과이다.

**진하게**, ~~취소~~, _이탤릭_ 

#### **들여쓰기**
들여쓰기를 하귀 위해서는 >를 이용하면 된다. 들여쓰기를 할수록 >의 개수를 늘려주면 된다 아래는 이를 나타내는 출력 결과이다.
>블럭 1단계
>>블럭2단계

####**목록**
index를 나타낼때 색칠 된 index를 나타낼 때는 *을 사용하고 빈 index를 나타낼 때는 +를 사용하면 된다. 또 이때 들여쓸때는 tab 을 사용하며 숫자 인덱스는 그냥 숫자로 나타내면 된다. 아래는 이를 이용한 예시이다.   
*사과
  * 딸기
    + 오렌지
  1. 참외
    + 수박

####**Code Block**
code block를 나타낼때는 ' ` ' 를 이용한다. 

####**하이퍼링크**
하이퍼링크를 하고싶은 부분에 대괄호 []를 해주고 옆에 소괄호를 이용하여 링크를 넣어준다. 아래는 이에 해당하는 예시이다.   
[링크](http://www.google.com)
