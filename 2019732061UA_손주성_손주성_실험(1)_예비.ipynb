###지도 학습(Supervised Learning)
지도 학습(Supervised Learning)은 문제와 정답을 모두 알려주고 공부시키는 방법이다.  
그래서 핵심인 정답을 알고 가르쳐 주는 전문가(Superviser)가 필요하다.  
또한, 사전에 정답을 아는 data를 이용해 반복적인 학습을 하고 학습을 바탕으로 새로운 data에 대해 정답을 예측하는 것이다.

#Markdown문법

확장자 명이 .md인 문서 파일이다.
coding을 통해서 문서를 작성하기 때문에 타이핑으로만 문서를 작성할 수 있다.

**타이틀**
 #의 갯수를 통해서 글자 크기를 조절할 수 있다. #의 갯수가 많아질수록 글자의 크기가 작아진다.  
ex)
###손주성
####손주성
#####손주성

**강조**  
 *, __, ~~, _ 를 통해서 4가지의 강조 효과를 줄 수 있다.  
Colab에서는 __이 진하게 로 되기 때문에 <,>,u를 사용해야한다.  
ex)  

**손주성**  
<u>손주성</u>  
~~손주성~~  
_손주성_  

**문장 줄바꿈**  
입력에서 Enter키를 통해 줄바꿈을 하더라도 실제로는 줄바꿈이 되지 않을 수 있다.
줄바꿈을 하고 싶으면 줄바꿈을 하고 싶은 문장 뒤에서 Spacebar을 2번 눌러주면 된다. 

ex)  
손 주 성  
손  
주  
성

**블럭(인용) 들여쓰기**  

꺽쇠(>)를 이용해서 블럭(인용) 들여쓰기를 할 수 있다.  
ex)  
>손주성  
>>손주성

**목록**  
 *, +, - 표시를 이용해서 목록을 표시할 수 있다. 검은색 동그라미로 목록이 표시가 된다. Tab키를 통해 들여쓰기 후 목록을 표시해주면 비어있는 동그라미로 목록을 표시할 수 있다.
ex)  
* 블랙핑크
  + 지수
  - 로제
  * 제니
  - 리사

**Code Block**  
텍스트 블록에 입력한 부분을 코딩할 때 쓰는 디자인으로 바꿔준다.  
 (`) 3개로 바꾸고 싶은 문장의 앞과 뒤에 두면 그 문장이 바뀐다.

ex)  
 ```   
import something
 a = 10
 print(a)
 if a>10:
    print(a)
```

**하이퍼링크**  
문구로 하이퍼링크를 표시해주고 싶으면 대괄호 안에 문구를 입력 후, 뒤에 괄호안에 링크를 넣어주면 된다.
문구가 필요없다면 꺽쇠를 통해서 바로 하이퍼링크를 만들 수 있다.

ex)  
[바로가기](http://google.com)  
<http://google.com>  

**수평선**  
 *, - 문자를 3개 이상 이어 쓰게 되면 수평선을 만들 수 있다.  

ex)  
***  
손주성  
___  

**표**  
 (|) 문자를 이용해 열을 구분할 수 있고, :을 통해서 정렬하고 싶은 방향을 설정할 수 있다.

**그림, 이미지**  
 ! 뒤에 대괄호에 이미지설명문구를 적은 후, 괄호안에 링크를 달아주면 이미지나 그림을 가져올 수 있다.  
ex)  
![이미지설명문구](https://search.pstatic.net/common/?src=http%3A%2F%2Fblogfiles.naver.net%2FMjAyMTA2MDhfMTY5%2FMDAxNjIzMTU4NDUwMjcx.QLVHOf3zY3DDT7R61T5HClgGiCo4GdimATqs3W3YFCkg.P5PP5dfNRnNF-lg6P3QfQWs4QPKOLfHf9vDIoPMyxK4g.JPEG.exia9133%2F20210608%25A3%25DF221939.jpg&type=a340)  

**체크박스**  
(todo list 같은 것에 좋다) 목록을 표시해 주는 문자 뒤에 대괄호를 통해 체크박스를 만들 수 있으며, 대괄호 안에 x 표시를 넣으면 체크된 체크박스를 만들 수 있다.  

ex)  
 * [ ] 비어있는 체크박스  
 * [x] 체크된 체크박스 

**수식**  
 $$ 표시 안에 Latex 수식을 넣으면 수식을 표시할 수 있다.  
자세한 Latex 수식은 [여기](https://cheris8.github.io/etc/MD-LaTex/) 에서 확인이 가능하다.  

ex)  
$\begin{matrix}1&2\\3&4\\ \end{matrix}$
$\begin{pmatrix}1&2\\3&4\\ \end{pmatrix}$
$\begin{bmatrix}1&2\\3&4\\ \end{bmatrix}$
$\begin{Bmatrix}1&2\\3&4\\ \end{Bmatrix}$
$\begin{vmatrix}1&2\\3&4\\ \end{vmatrix}$  
  
(&로 열을 구분, \\로 행을 구분)
