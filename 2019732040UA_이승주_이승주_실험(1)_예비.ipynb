## **강의영상2 요약**

* **Goodle Colab**

 * 구글 코랩은 웹 브라우저에서 텍스트와 코드를 자유롭게 오가며 작성할 수 있는 온라인 에디터이다. 웹 브라우저 내에서 다양한 머신러닝 기법을 프로그래밍한 후 즉석에서 그 결과 또한 확인할 수 있다.주로 텍스트 셀, 코드 셀로 나뉘며 GPU를 지원하여 제한된 범위 내에서는 개인 PC보다도 빠른 성능을 보여주기도 한다. 구글 드라이브 계정과 연결하여 사용하면 더욱 편리하게 이용할 수 있다.


- **Jupyter Notebook (Markdown)**

 *  Jupyter Notebook은 Markdown이라는 문법을 이용해 문서와 소스코드를 한 문서에 배치에 작업할 수 있는 파일 형식이자 플랫폼이다. 소스코드와 텍스트는 각각 '블록' 단위로 문서 내에 배치할 수 있다. 소스코드는 해당 프로그래밍 언어 문법을, 텍스트는 Markdown 문법을 따른다.
 요약하면, Markdown 문법은 텍스트 문서를 마치 프로그래밍하듯 소스코드를 입력해 그 컴파일 결과가 정리된 문서로 보여지는 것이라 생각할 수 있다.
