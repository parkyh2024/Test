###1.머신러닝이란?
* 지도 학습(Supervised Learning)  
문제와 정답을 모두 알려주고 공부시키는 방법  
>지도학습 알고리즘  
  - [x] *분류기법(Classification techniques)* : 입력 데이터를 식별하기 위해 알고리즘에 이산 값을 예측하도록 요청한다. 보통 이미지 인식에 사용된다.  
  - [x] *선형 회기(Linear regression)* : 연속 데이터를 사용해서 값을 예측한다.

* 비지도 학습(Unsupervised Learning)  
답을 가르쳐주지 않고 공부시키는 방법  
* 강화 학습(Reinforcement Learning)  
보상을 통해 상은 최대화, 벌은 최소화하는 방향으로 행위를 강화하는 학습

###3. Jupyter Notebook(Markdown)
`Jupyter Notebook`은 `Markdown`이라는 문법을 이용해 <u>문서와 소스코드를 한 문서</u>에 작업할 수 있는 파일 형식이자 플랫폼이다.  
>**마크다운(Markdown)이란?**  

`HTML`경험이 없는 사람도 누구나 쉽게 헤더, 글 머리 기호, 이미지 삽입, 링크 등 다양한 서식을 추가하는 방식의 문서 편집 문법입니다.  

>**다양한 서식 사용법**  

1. 타이틀 : `# ` 
#title (#1개)  
##title (#2개)

2. 강조  
<img src="https://img1.daumcdn.net/thumb/R1280x0/?scode=mtistory2&fname=https%3A%2F%2Fblog.kakaocdn.net%2Fdn%2FsCfkC%2FbtraRwIWWGl%2Fc4eWD9P94kUi1laMdSTNPK%2Fimg.jpg" width="600" height="100">  

3. 문장 줄바꿈 : `문장 끝에 space 2개`   

4. 들여쓰기 : `>` 를 많이 쓸수록 안쪽에 표시
>`>` 1개
>>`>>` 2개  

5. 목록(List)구성  
* `*` 입력
  * `(tab)*`입력  

6. 링크  
`[대표글](url)`

7. 이미지  
`![이미지 설명](이미지 url)`

8. 수평선  
`---` or `***`  

9. 코드 블럭  
* 한줄 코드 블럭은  ``사이에 글 입력  
* 여러줄 코드 블럭은 ` ``` 와 ``` `사이에 글 입력  

10. 수식 표현(Latex)  
* `$`로 수식의 앞 뒤를 감싸면 수식을 작성할 수 있음(기본은 왼쪽 정렬)  
* `$$`로 수식의 앞 뒤를 감싸면 중앙 정렬  
* `aligned` 심볼을 사용하면 특정 문자를 기준으로 정렬  
* 분수 표현법 `/over` or `/frac`
* 절댓값 표현 `|`
* 극한 표현 `\lim`, 시그마 표현 `\sum`
* 벡터 표현 `\vec` 
* 행렬 표현 `matrix` 심볼사용
