## **Notebook 단축키**
여러가지 단축키가 존재하지만, 그 중에서도 자주 사용하는 것을 정리하면 다음과 같다.  
* 실행 후 다음 셀로 이동 : Shift + Enter
* 저장 : Ctrl + S
* 마크다운으로 전환 : Ctrl + M M
* 코드셀에 줄번호 부여 : Ctrl + M L

## **그 외 기능들**
### **타이틀** : 문자 '#'의 개수를 다르게 하여 타이틀을 정할 수 있다.
### **강조** : 텍스트 앞 뒤로 문자를 삽입함으로써 해당 텍스트를 강조할 수 있다.
### **줄 바꿈** : 텍스트 뒤에 스페이스바를 두 번 입력하면 문장 줄바꿈을 할 수 있다.
### **블럭 들여쓰기** : 텍스트 앞에 꺽새표시를 추가하면 해당 줄을 들여쓰기할 수 있다.
### **목록** : 텍스트 앞에 *, +, - 등을 추가하면 색이 칠해진 동그라미가 추가되어 목록화될 수 있다. 반면에, 단지 tab을 누르게 된다면 들여쓰기가 된 상태에서 색이 칠해지지 않은 동그라미가 추가되면서 목록화된다.
### **Code Block** : 텍스트 앞 뒤에 ```을 추가하면 해당 텍스트를 파이선과 같은 것으로 코드화시킬 수 있다. 
### **하이퍼링크** : 해당 텍스트를 대괄호로 묶고, 그 옆에 소괄호로 링크 주소를 입력하면 하이퍼링크를 활성화할 수 있다.
### **수평선** : '-'을 3개 이상 입력하면 수평선을 추가할 수 있다.
### **표** : 집적 입력할 수 있지만, 내용이 많은 경우 시중에 나와 있는 사이트를 참고하여 입력한다. 
### **그림** : 하이퍼링크와 방식이 동일하지만, 앞에 느낌표를 추가해야한다.
### **체크박스** : 텍스트 앞에 빈 대괄호를 입력하면 비어있는 체크박스를, 대괄호 안에 x를 입력하면 체크된 체크박스를 추가할 수 있다.
### **수식** : 웬만한 수식을 입력할 수 있고, 이는 Latex 수식을 참고하도록 한다.
